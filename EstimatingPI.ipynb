Task 1: Estimating $\pi$



```python
# Estimating PI
from random import random

import jax
import jax.numpy as jnp


def pi_python():
  npoints = 100_000
  xy = [(random(), random()) for _ in range(npoints)]

  inside = 0
  for x, y in xy:
    if x**2 + y**2 < 1:
      inside += 1.0

  return inside * 4.0 / npoints


print(f"pi ~= {pi_python()}")

print("\nPure Python:")
%timeit -n 10 pi_python()
```
